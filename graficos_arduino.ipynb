```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("monitor.csv.txt", sep=';')
print(df.head())
print(df.tail())
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Tempo'], df['TotalWh'], label='Consumo Total (Wh)', color='black', linewidth=2)
plt.title('Evolução do consumo Total ao Longo do Tempo')
plt.xlabel('Tempo')
plt.ylabel('Consumo (Wh)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
consumo_final = df[['CaixaSomWh', 'VentiladorWh', 'TVWh', 'GeladeiraWh']].iloc[-1]

plt.figure(figsize=(8,5))
consumo_final.plot(kind='bar', color=['orange', 'blue', 'green', 'red'])
plt.title('Consumo Total de Cada Aparelho (Wh)')
plt.ylabel('Consumo (Wh)')
plt.xlabel('Aparelho')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(consumo_final, labels=consumo_final.index, autopct='%1.1f%%', startangle=90)
plt.title('Participação de Cada Aparelho no Consumo Total')
plt.show()
```
